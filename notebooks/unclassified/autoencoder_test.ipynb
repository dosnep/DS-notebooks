```python
import numpy as np 
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
class AutoEncoder(keras.Model):
    def __init__(self):
        super().__init__(name='auto_encoder')
        self.__l1 = keras.layers.Dense(100, activation='relu')
        self.__l2 = keras.layers.Dense(50, activation='relu')
        self.__l3 = keras.layers.Dense(40, activation='relu')
        self.__l4 = keras.layers.Dense(50, activation='relu')
        self.__l5 = keras.layers.Dense(100, activation='relu')        

    def call(self, input_data):
        print("This is a Deep Learning Auto Encoder")
        x = self.__l1(input_data)
        x = self.__l2(x)
        x = self.__l3(x)
        x = self.__l4(x)
        return self.__l5(x)
    
    def encoder(self, input_data):
        print("Encoder")
        x = self.__l1(input_data)
        x = self.__l2(x)
        return self.__l3(x)
```

```python
data = np.random.uniform(0, 1, 100000)
data = data.reshape(10, 100, 100)
encoder = AutoEncoder()
encoder.compile(optimizer='adam', loss='binary_crossentropy')
encoder.fit(data, data, epochs=50)
```

```python
encoder_data = encoder.encoder(data)
predictions = encoder.predict(data)
```

```python
figure = plt.figure(figsize=(20, 10))
ax = plt.subplot(131)
ax.imshow(data[1], cmap='gray')
#
ax.grid(True)
ax = plt.subplot(132)
ax.imshow(encoder_data[1], cmap='gray')
ax.grid(True)
ax = plt.subplot(133)
ax.imshow(predictions[1], cmap='gray')
ax.grid(True)
```
